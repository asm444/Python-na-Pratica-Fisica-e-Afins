 O laço for

O **for** é uma das ferramentas mais úteis do Python (minha preferida). Ele é útil para fazer cálculos longos, porém repetitivos. Mostra claramente o poder da computação. A lógica é essencialmente:

**for CONTADOR em INTERVALO: \\
  INSTRUÇÕES**

Vejamos agora alguns exemplos.

 




# Exemplos de laços de repetição

```python
for i in range(7):    # primeiro definimos um contador i, que irá rodar em um intervalo de números. É usual definir tais intervalos usando range()
  print(2)            # printamos o número dois 7 vezes na tela. Interessante, não?
```

O laço **for ( )** é particularmente útil para gerar **sequências** de números. Vejamos agora alguns exemplos.

```python
for i in range(5):    # contador definido de maneira análoga ao anterior
  print(2*i)          # agora a instrução é printar múltiplos de 2 para cada valor do contador
```

```python
for i in range(13):    # contador definido de maneira análoga ao anterior
  print(2**i)          # agora a instrução é printar potências de 2 para cada valor do contador
```

Vamos agora ver que podemos usar esse laço de repetição pra inserir elementos em listas a cada iteração. Fazemos isso primeiro definindo uma lista vazia e usando o comando **append( )** dentro do laço.

```python
power = []                  # definição de uma lista vazia
for i in range(1000):       # contador
  power.append(2**i)        # cada vez que o contador roda,  uma potência de 2 é adicionada a nossa lista

print(power)                # printa nossa lista
#print(power[len(power)-1]) # printa o último número, que é enorme
```

```python
# Fatorial

fatorial = []
p = 1
for i in range(7):
  p = p*(i+1)
  fatorial.append(p)

print(fatorial)                   # lista cujos elementos são os respectivos fatoriais de (i+1)
```

```python
fatorial = []
p = 1
for i in range(10**3):
  p = p*(i+1)
  fatorial.append(p)

print(fatorial[len(fatorial)-1])
```

# Atividade: cálculo de π

Vamos agora ver como calcular o famoso $\pi$ usando laços de repetição. Usaremos que ele é definido como a razão entre o perímetro $p$ da circunferência e seu diâmetro $d=2r$ :

$$ \pi = \frac{p}{d}\ , $$

para isso, vamos aproximar uma circunferência por um polígono com um número grande de lados 

** INSERIR VÍDEO **

e calcular a razão entre o perímetro do polígono (soma de seus lados) sob o diâmetro da circunferência na qual ele está inscrito.



O lado $l$ de um perìmetro regular de $n$ lados pode ser calculado usando

\begin{equation}
  l = r\sqrt{2}\sqrt{1-\cos \theta_{n}} \, , 
\end{equation}

em que $\theta_n = 360º/n$ é o ângulo do polígono regular.

** VOU DERIVAR A FÓRMULA NO MINICURSO E DEPOIS COLOCAR AQUI NAS NOTAS **

```python
n = 100                                               # número total de iterações
r = 3                                                 # raio(arbitrário) da circunferência
pi = []                                             # definição de uma lista vazia
for i in range(n):
  l = r*np.sqrt(2)*np.sqrt(1-np.cos((2*np.pi)/(i+3))) # fórmula para cálculo do lado do polígono de n lados. O contador aparece com i+3 porque o menor polígono é o triângulo, de 3 lados
  p = (i+3)*l                                         # perímetro
  lim = (p)/(2*r)                                     # razão entre o perímetro e o raio
  pi.append(lim)                                      # lista com as razões entre os perímetros dos polígonos e o raio da circunferência

print(pi)                                             # printa a sequência de iterações
print(pi[-1])                                        # printa o último termo da sequência, que é o valor aproximado de pi
```

# Polígonos

```python
import matplotlib.pyplot as plt
import math as mt

def ang(m):                           # definimos o ângulo do polígono
  return mt.radians(360/m)            # essa linha converte de graus pra radianos

n = 7                                 # vamos começar gerando um polígono de 7 lados
x0 = [1]      
y0 = [0]                              # o ponto inicial é (1,0), suas coordenadas iniciais são x0 = 0, e y0 = 1
for i in range(n):                    # gera pontos sob a circunferência unitária espaçados com o ângulo do polígono
  x0.append(np.cos((i+1)*ang(n)))     # coordenadas x dos pontos
  y0.append(np.sin((i+1)*ang(n)))     # coordenadas y dos pontos

x = np.linspace(-1,1,1000)            # intervalo (-1,1)
yp = np.sqrt(1-x**2)                  # parte de cima da circunferência
yn = -yp                              # parte de baixo da circunferência

plt.figure(figsize=(7,7))             # tamanho do gráfico
plt.plot(x0,y0, color='red')          # polígono de vermelho
plt.plot(x,yp, color='blue')          # parte de cima da circunferência, de azul
plt.plot(x,yn, color='blue')          # parte de baixo da circunferência, de azul
plt.show()  

#print(coordd)
```

Agora que entendemos a lógica para gerar polígonos, vamos criar uma função que resume a sintaxe acima, generalizando à para valores arbitrários de $n$.

```python
def polygon(n):                           # essa função gera pontos sob a circunferência associada aos vértices de um polígono de n lados 
    x0 = [1]                              # a função foi definida simplesmente repetindo a sintaxe acima
    y0 = [0]
    for i in range(n):
        x0.append(np.cos((i+1)*ang(n)))
        y0.append(np.sin((i+1)*ang(n)))
    return x0,y0                          # a função retorna um par de listas: uma com as coordenadas x dos pontos e outra as coordenadas y

print(polygon(3)[0])                      # a 0-ésima entrada fornece as coordenadas x dos pontos
print(polygon(3)[1])                      # a 1-ésima entrada fornece as coordenadas y dos pontos
```

Testando nossa função:

```python
plt.figure(figsize=(7,7))
plt.plot(x,yp, color='blue')
plt.plot(x,yn, color='blue')
plt.plot(polygon(6)[0], polygon(6)[1], color = 'red')
plt.show() 
```

Agora podemos também usar laços de repetição para plotar vários gráficos seguidos de polígonos. Vamos então colocar a sintaxe acima dentro de um **for** e variar o número de lados com um contador.

```python
m = 10                                                      # m = 10 iterações    
for i in range(m):
    plt.figure(figsize=(7,7))
    plt.plot(polygon(i+3)[0],polygon(i+3)[1], color='red')  # aqui coloquei i+3 lados porque o polígono com menor número de lados é o triângulo, com i=0
    plt.plot(x,yp, color='blue')
    plt.plot(x,yn, color='blue')
    plt.show() 
```

# Vídeos com Polígonos

Logo acima, vimos como desenhar polígonos em sequência usando laços de repetição. Agora vamos ver como fazer vídeos, e a lógica seguida será **exatamente a mesma acima**. 

As animações são feitas essencialmente da mesma forma que os desenhos em stop motion. Precisamos apenas de algo que junte todos os gráficos em um único vídeo. quem faz isso é a biblioteca **animation** do matplotlib. 

```python
import matplotlib.animation as animation # importa a biblioteca de animação
```

Vamos agora definir nossa conjunto de gráficos. Para fazer o vídeo, precisamos fazer isso usando o comando subplots. Também precisamos gerar vários gráficos de uma vez, porém essa é a parte fácil. Vamos gerar $m=30$ gráficos.

```python
fig1, ax =  plt.subplots(figsize=(7,7))                                           # subplots na fig1

m = 30                                                                            # número de "frames" do vídeo
pols = []                                                                         # lista vazia que se tornará uma lista de plots
for i in range(m):
  #cap = str(i+3) - legenda, não consegui variar com o contador
  #ax.set_xlabel(cap, fontsize=15)
  poli, =  ax.plot(polygon(i+3)[0], polygon(i+3)[1], linestyle='-', color='red')  # plots do polígono de i+3 lados
  circp, = ax.plot(x, yp, linestyle='-', color='blue')                            # plots da parte de cima da circunferência
  circn, = ax.plot(x, yn, linestyle='-', color='blue')                            # plots da parte de baixo da circunferência
  pols.append([poli,circp,circn])                                                 # lista com os plots
```

```python
ani1 = animation.ArtistAnimation(fig1, pols, interval=700) # Esse comando cria a animação. Ele toma como inputs uma figura, uma lista de gráficos, e um intervalo para fazer o vídeo
plt.rcParams['animation.html'] = 'html5' 
ani1
```
