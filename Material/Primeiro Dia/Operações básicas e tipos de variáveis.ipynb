# Tipos de variáveis

Variáveis são nomes que damos pra alguma informação, como por exemplo, quando dizemos

$$ x = 2 $$

estamos declarando uma variável. Elas vêm em vários tipos diferentes, e cada um carrega um tipo de informação diferente. Vamos agora ver um pouco sobre os tipos de variáveis mais comuns.

## 1. str - string

Variáveis do tipo string carregam informações como palavras ou textos. Por exemplo:



```python
nome = 'luiz' # declarar com aspas!
nome
```

Também podemos fazer ela aparecer na tela usando o comando print(**VARIÁVEL**) . Por exemplo:

```python
sobrenome = 'demétrio'
print(sobrenome)
```

Note que 

```python
nome
sobrenome
```

é diferente de 

```python
print(nome)
print(sobrenome)
```

Com o primeiro caso printando apenas a última variável, e o outro sendo útil quando temos mais de uma variável.

## 2. int - inteiro

Variáveis do tipo inteiro carregam números inteiros. Por exemplo:

```python
idade = 23
print(idade)
```

Agora que temos dois tipos diferentes de variáveis, é útil termos algum jeito de compará-las. Uma forma de fazer isso é usando o comando type(**VARIÁVEL**). Vejamos alguns exemplos:

```python
type(nome)
```

```python
print(type(nome))
print(type(idade))
```

Ao analisarmos programas de outras pessoas, é útil usar esse comando para verificar os tipos de variáveis desconhecidas. Obviamente, nesse caso poderia-se ver facilmente que os tipos eram diferentes. Porém, veremos agora um outro tipo de variável em que a diferença é menos trivial.

## 3. float - número real

Variáveis do tipo float representam números reais. O termo "número de ponteiro flutuante" é usado devido à posição da vírgula importar nesse tipo de número:

$$ 3,42 \neq 34,2 \, .$$

Vejamos agora alguns exemplos.

```python
altura = 1.82 # o Python segue o jeito americano de escrever números, usando pontos ao invés de vírgulas
print(altura)
print(type(altura))
```

Note que esse tipo é diferente do int, sendo útil ter o comando type( ) para diferenciá-los.

# Operações com variáveis 

Veremos agora como operar com variáveis, o que facilita muito a escolha de códigos.

```python
a = 2 
b = 3 # acabamos de declarar nossas variáveis

print(a+b) # soma
print(a-b) # subtração
print(a*b) # multiplicação com *
print(a/b) # divisão
print(a**b) # potenciação
```

Note que podemos definir novas variáveis com base nelas:

```python
soma = a+b 
subtração = a-b 
produto = a*b
divisão = a/b
potenciação = a**b
```

```python
print(soma)
print(subtração)
print(produto)
print(divisão)
print(potenciação)
```

Note que, ao printarmos os tipos dessas variáveis, obtemos:

```python
print(type(soma))
print(type(subtração))
print(type(produto))
print(type(divisão))
print(type(potenciação))
```

Veja que o tipo mudou de int para float ao fazermos a divisão, porque a divisão de dois naturais nem sempre é um natural.

E, é claro, sempre podemos fazer a mesma coisa para números absurdamente grandes:

```python
c = 8758757684567847864675876566785
d = 3478689686775764674657456565676

print(c+d)
print(c-d)
print(c*d)
print(c/d)
```

```python
numeros = [1,2,3] # definimos listas usando colchetes
print(numeros)
print(type(numeros))
```

Também podemos criar listas usando variáveis antigas:

```python
dados = [nome,idade,altura]
print(dados)
```
