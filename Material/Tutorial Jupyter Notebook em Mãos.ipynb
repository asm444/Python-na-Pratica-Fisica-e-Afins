# Jupyter Notebook

<p>O Jupyter Notebook é uma ferramenta poderosa que possíbilita escrever um código em python de forma interativa a cada seção de código através de elementos chamados células e saídas, da mesma maneira possibilita inserir imagens, formulas, gráficos e instruções. Isso permite uma fluidez no código e a contrução desde uma história do código quanto a um tutorial. Essa maneira de contruir um código permite a facil compreensão de dados, códigos e tirar uma conclusão clara sobre o que está sendo executado, o que não acontece na leitura de um código em editor de texto, pois muitas vezes acaba se tornando uma investigação e demanda mais tempo para um trabalho.  </p>

## Tutorial Prático do Jupyter Notebook

### Elementos do Jupyter Notebook

### Estrutura de células no Jupyter Notebook:

Existem três tipos de células no Jupyter: **código**, **markdown** e **raw NBconvert**.
- As **células de código** são aquelas que permitem a execução de códigos de uma determinada linguagem (neste caso, Python). 
- As **células do tipo markdown** permitem a utilização da linguagem de marcação **Markdown**, que permite a edição de textos. 
- Já as **raw NBconvert** permitem a execução de códigos externamente, tais como o LaTeX.

Ao executar `%matplotlib inline` em uma das células (só é necessário executá-lo uma vez)

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.plot(range(0,10), [x**2 for x in range(0,10)])
```

O comando `%matplotlib notebook` permite a criação de uma gráfico interativo, parecido com a janela nativa do Matplotlib:

```python
%matplotlib notebook
plt.plot(range(0,10), [x**2 for x in range(0,10)])
```
