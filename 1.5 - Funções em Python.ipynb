```python
def check_sinal (x):
    if x>0:
        print("Positivo.")
    elif x==0:
        print("Nulo.")
    else:
        print("Negativo.")
```

```python
def Soma_sequencia(sequencia):#sum(lista)
    valor_total_da_soma = 0
    for elemento in sequencia:
        valor_total_da_soma += elemento
    return valor_total_da_soma
```

## Multiplicação de Matrizes

$$\displaystyle A_{mn} \cdot B_{np} = C_{mp}$$

```python
#o objeto matriz não exite no python

#vamos criar usando listas dentro de listas
#Uma lista contendo uma coleção de outras listas

#A quantidade de listas dentro da lista pode ser defindo como a quantidade de linhas da matriz.
#A quantidade de elementos dentro de cada lista pode ser definido como a quantidade de colunas da matriz.

#exemplos
'''
[[-1,3],
 [4,2]]

[[1,0,0],
 [0,1,0],
 [0,0,1]]'''

matriz_A = [[1,0,0],
            [0,1,0],
            [0,0,1]]




def multiplicação_de_matrizes(matriz1, matriz2):
    
    if len(matriz1)!=len(matriz2[0]):
        print("Não é possível multiplicar as matrizes.")
        print("O número de colunas da primeira matriz é diferente do número de linhas da segunda matriz.")
        
    
    else:
        # O número de colunas da primeira matriz é dado por len(matriz1)
        # O número de linhas da segunda matriz é dada por len(matriz2[0])
        
        matriz_resultante = []
        
        # O número de colunas da segunda matriz é dado por len(matriz2)
        
        for coluna in range(len(matriz2)):
            #Adicionando colunas
            matriz_resultante.append([])
        
            
        # Manipulando um elemento genérico da matriz C_m_n = A_m_p x B_p_n
        
        for m in range(len(matriz1[0])): # Multiplicando linha 
            
            for n in range(len(matriz2)): # Por coluna             
                elemento = 0
                c =[]
                for p in range(len(matriz1)):
                    
                    elemento = matriz1[m][p] * matriz2[p][n]
                    c.append(elemento)

                elemento_resultante = Soma_sequencia(c)
                    
                matriz_resultante[m].append(elemento_resultante)
        print('Matriz Resultante: \n')        
        for m in range(len(matriz1[0])):
            print('|',end=' ')
            for n in range(len(matriz2)):
                if n<len(matriz2):
                    print(matriz_resultante[m][n], end=' ')
            print('|')


multiplicação_de_matrizes(matriz_A,matriz_A)
                   
```
